Importing necessary python libraries


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import LabelEncoder
```

importing dataset

```python
df = pd.read_csv("IRIS.csv")
```

Displaying the first few rows and summary statistics

```python
print(df.head())
```

```python
print(df.describe())
```

```python
print(df.info())
```

Setting up the matplotlib figure for feature distribution

```python
plt.figure(figsize=(12, 8))
```

Plotting histograms for each feature

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[feature], kde=True, bins=20)
    plt.title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()
```

Computing the correlation matrix

```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

 Setting up the matplotlib figure for the heatmap

```python
plt.figure(figsize=(10, 8))
```

 heatmap

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Features')
plt.show()
```

Pair plot of the features colored by species

```python
sns.pairplot(df, hue='species', diag_kind='kde')
plt.suptitle('Pair Plot of Features Colored by Species', y=1.02)
plt.show()
```

Count plot for species distribution

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='species')
plt.title('Species Distribution')
plt.show()
```
